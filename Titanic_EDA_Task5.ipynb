```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set(style="darkgrid")

print("Libraries loaded.")
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
print("Before Number of Null values:\n",df.isnull().sum())
```

```python
# Shape and columns
print("Shape:", df.shape)
print("Columns:", df.columns)

# Basic info
df.info()

# Summary statistics
df.describe()

# Missing values
df.isnull().sum()
```

```python
#age distribution
sns.histplot(df['Age'].dropna(), kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#Survival Count
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
#Survival by Sex
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival by Sex")
plt.show()
```

```python
# Age vs Fare
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare by Survival")
plt.show()
```

```python
#Multivariate Analysis
sns.pairplot(df[['Age', 'Fare', 'Survived', 'Pclass']], hue='Survived')
plt.suptitle("Pairplot of Age, Fare, Pclass and Survival", y=1.02)
plt.show()
```

```python
# Correlation Heatmap
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Boxplots
sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.show()

sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# Fill missing Age with median (safe method)
df['Age'] = df['Age'].fillna(df['Age'].median())

# Log transform Fare
df['Fare_log'] = np.log1p(df['Fare'])

# Plot the log-transformed fare
sns.histplot(df['Fare_log'], kde=True)
plt.title("Log Transformed Fare")
plt.show()
```
